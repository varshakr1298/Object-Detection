## Cat face Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('cat.jpg')
image=cv2.resize(image,(500,500))
face_cascade=cv2.CascadeClassifier('haarcascades/haarcascade_frontalcatface_extended.xml')
#cv2.imshow('smile',image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
# Converting the image into grayscale for computation
import cv2
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray,1.1,15)

simage=image.copy()
for (x,y,w,h) in faces:
    cv2.rectangle(simage,(x,y),((x + w),(y + h)),(255,0,0),3)              
```

```python
simage=cv2.cvtColor(simage,cv2.COLOR_BGR2RGB)
plt.imshow(simage)
plt.show()
```
